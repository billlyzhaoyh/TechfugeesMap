```python
import pandas as pd
```

```python
map_data=pd.read_csv('map_data.csv')
```

### Global scale

```python
import folium
```

```python
global_data=map_data.loc[map_data['SettlementType']=='- Country Overview',:]
```

```python
#TBC is not plottable so we remove that one from the frame
```

```python
global_data_to_plot=global_data[global_data.Cases != 'TBC']
```

```python
global_data_to_plot
```

```python
dict_to_json={
    'GREECE / COUNTRY OVERVIEW':'GRC',
    'LEBANON / COUNTRY OVERVIEW':'LBN',
    'ðŸ‡ºðŸ‡¸UNITED STATES / COUNTRY OVERVIEW':'USA',
    'Afghanistan Overview':'AFG'
}
```

```python
global_data_to_plot['CountryCode']=global_data_to_plot['Name'].apply(lambda x: dict_to_json[x])
```

```python
global_data_to_plot=global_data_to_plot.astype({'Cases': 'int','CountryCode':'string'})
```

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
country_geo = f'{url}/world-countries.json'
```

```python
country_geo
```

```python
country_codes=global_data_to_plot['Name'].apply(lambda x: dict_to_json[x]).tolist()
```

```python
import urllib.request, json 
with urllib.request.urlopen(country_geo) as url:
    data = json.loads(url.read().decode())
```

```python
trimmed_data={}
```

```python
data['type']
```

```python
list_of_features=[]
for key,value in data.items():
    if key=='features':
        for country in value:
            if country['id'] in country_codes:
                list_of_features.append(country)
```

```python
for key,value in data.items():
    if key=='features':
        trimmed_data['features']=list_of_features
    else:
        trimmed_data[key]=value
```

```python
bins = list(global_data_to_plot['Cases'].quantile([0, 0.1, 0.9,1]))
```

```python
m = folium.Map(location=[48, -102], zoom_start=2)
folium.Choropleth(
    geo_data=trimmed_data,
    name='choropleth',
    data=global_data_to_plot,
    columns=['CountryCode', 'Cases'],
    key_on='feature.id',
    fill_color='YlOrRd',
    fill_opacity=0.5,
    line_opacity=0.5,
    legend_name='Number of confirmed COVID cases',
    bins=bins,
    reset=True
).add_to(m)

folium.LayerControl().add_to(m)
```
